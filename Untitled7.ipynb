```python
# Step 3: Patch main_HTNet.py — change epochs from 800 to 50
# We also patch the hardcoded 'cuda:0' to be device-safe

with open('main_HTNet.py', 'r') as f:
    src = f.read()

# Change epoch count
src = src.replace('epochs = 800', 'epochs = 50')

# Fix hardcoded cuda:0 — use dynamic device instead
src = src.replace(
    "device='cuda:0'",
    "device=('cuda' if torch.cuda.is_available() else 'cpu')"
)

with open('main_HTNet.py', 'w') as f:
    f.write(src)

# Confirm the patch
!grep -n "epochs" main_HTNet.py
!grep -n "device=" main_HTNet.py | head -5
print("\n✅ Patched: epochs=50, device is now dynamic")
```

```python
# Step 4: Run training — using the repo's own script
# We capture stdout so we can parse UF1/UAR per subject for plotting
import subprocess, sys, re, time

start = time.time()

result = subprocess.run(
    [sys.executable, 'main_HTNet.py', '--train', 'True'],
    capture_output=True, text=True
)

training_log = result.stdout
print(training_log[-5000:])   # print last 5000 chars (tail of log)
if result.stderr:
    print("STDERR:", result.stderr[-2000:])

print(f"\nTotal time: {(time.time()-start)/60:.1f} min")
```

```python
import re, numpy as np

# Parse 'best UF1: X.XXXX | best UAR: X.XXXX' lines from log
uf1_vals = re.findall(r'best UF1: ([\d.]+)', training_log)
uar_vals = re.findall(r'best UAR: ([\d.]+)', training_log)

uf1_series = [float(v) for v in uf1_vals]
uar_series = [float(v) for v in uar_vals]

print(f"Subjects processed : {len(uf1_series)}")
if uf1_series:
    print(f"Final UF1          : {uf1_series[-1]:.4f}  (paper 800ep: 0.8603)")
    print(f"Final UAR          : {uar_series[-1]:.4f}  (paper 800ep: 0.8475)")
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('HTNet — Cumulative UF1 / UAR as LOSO subjects are processed (50 epochs)',
             fontsize=13, fontweight='bold')

for ax, vals, metric, paper_val in zip(
        axes,
        [uf1_series, uar_series],
        ['UF1', 'UAR'],
        [0.8603, 0.8475]):
    ax.plot(range(1, len(vals)+1), vals, 'o-', color='royalblue', linewidth=2, markersize=5)
    ax.axhline(paper_val, color='red', linestyle='--', linewidth=1.5,
               label=f'Paper (800ep) = {paper_val}')
    ax.set_xlabel('Subjects processed (LOSO fold)', fontsize=11)
    ax.set_ylabel(metric, fontsize=11)
    ax.set_title(f'Cumulative {metric}', fontsize=11, fontweight='bold')
    ax.legend(fontsize=10); ax.grid(alpha=0.3)
    if vals:
        ax.annotate(f'Ours: {vals[-1]:.4f}',
                    xy=(len(vals), vals[-1]),
                    xytext=(-40, 10), textcoords='offset points',
                    fontsize=10, color='royalblue', fontweight='bold')

plt.tight_layout()
plt.savefig('loso_progress.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
from IPython.display import display

final_UF1 = uf1_series[-1] if uf1_series else 0.0
final_UAR = uar_series[-1] if uar_series else 0.0

methods  = ['LBP-TOP','Bi-WOOF','AlexNet','GoogLeNet','VGG16',
            'OFF-ApexNet','STSTNet','CapsuleNet','Dual-Inception',
            'EMR','RCN','FeatRef','SLSTT-LSTM',
            'HTNet (Paper 800ep)', 'HTNet (Ours 50ep)']
full_uf1 = [.5882,.6296,.6933,.5573,.6425,.7196,.7353,.6520,.7322,
            .7885,.7432,.7838,.8160,.8603, final_UF1]
full_uar = [.5785,.6227,.7154,.6049,.6516,.7096,.7605,.6506,.7278,
            .7824,.7190,.7832,.7900,.8475, final_UAR]
smic_uf1 = [.2000,.5727,.6201,.5123,.5800,.6817,.6801,.5820,.6645,
            .7461,.6326,.7011,.7400,.8049,'—']
smic_uar = [.5280,.5829,.6373,.5511,.5964,.6695,.7013,.5877,.6726,
            .7530,.6441,.7083,.7200,.7905,'—']
casme_uf1= [.7026,.7805,.7994,.5989,.8166,.8764,.8382,.7068,.8621,
            .8293,.8512,.8915,.9010,.9532,'—']
casme_uar= [.7429,.8026,.8312,.6414,.8202,.8681,.8686,.7018,.8560,
            .8209,.8123,.8873,.8850,.9516,'—']
samm_uf1 = [.3954,.5211,.6104,.5124,.4870,.5409,.6588,.6209,.5868,
            .7754,.7601,.7372,.7150,.8131,'—']
samm_uar = [.4102,.5139,.6642,.5992,.4793,.5392,.6810,.5989,.5663,
            .7152,.6715,.7155,.6430,.8124,'—']

df = pd.DataFrame({'Method':methods,
    'Full UF1':full_uf1,'Full UAR':full_uar,
    'SMIC UF1':smic_uf1,'SMIC UAR':smic_uar,
    'CASME II UF1':casme_uf1,'CASME II UAR':casme_uar,
    'SAMM UF1':samm_uf1,'SAMM UAR':samm_uar})

display(df.style.set_caption('Table 2 — UF1/UAR Comparison (LOSO)'))

# Bar chart — Full composite only
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Table 2 — Full Composite UF1 / UAR Comparison', fontsize=13, fontweight='bold')
colors = ['#4878CF']*13 + ['#2ca02c'] + ['#FF9500']
for ax, vals, metric, paper_best in zip(
        axes, [full_uf1, full_uar], ['UF1','UAR'], [0.8603, 0.8475]):
    bars = ax.bar(range(len(methods)), vals, color=colors, edgecolor='white')
    ax.set_xticks(range(len(methods)))
    ax.set_xticklabels(methods, rotation=50, ha='right', fontsize=8)
    ax.set_ylabel(metric); ax.set_ylim(0.3, 1.05)
    ax.axhline(paper_best, color='red', linestyle='--', linewidth=1.2,
               label=f'Paper best={paper_best}')
    ax.legend(fontsize=8); ax.grid(axis='y', alpha=0.3)
    ax.set_title(f'Full Composite {metric}', fontweight='bold')

plt.tight_layout()
plt.savefig('table2_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```
